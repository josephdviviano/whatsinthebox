```python
hatelist = []
with open(hatecsv, newline='') as inputfile:
    for row in csv.reader(inputfile):
        hatelist.append(row[0])
print(len(hatelist))
```

```python
hatebig=[]
hateuni=[]
for hate in hatelist:
    if len(hate.split(' ')) == 2 :
        hatebig.append(tuple(hate.split(' ')))    
    elif len(hate.split(' ')) == 1 :
        hateuni.append(hate)
```

```python
#Simple word search
start = time.time()
pornsites={}
for doc in docs:
    clean = [w.strip(string.digits + string.punctuation).lower() for w in word_tokenize(doc.content) if w not in stop_words]
    pwds=len([w for w in (word_tokenize(str(' '.join(clean)))) if w in hateuni])
    pornsites[doc.url]=pwds
duration=str(datetime.timedelta(seconds=time.time() - start))
print("Number of pages counted: ", len(pornsites))
print("Duration: ", duration)
```

```python
#Simple word search + debugging
start = time.time()
pornsites={}
for doc in docs:
    clean = [w.strip(string.digits + string.punctuation).lower() for w in word_tokenize(doc.content) if w not in stop_words]
    pwds=set([w for w in (word_tokenize(str(' '.join(clean)))) if w in hateuni])
    pornsites[doc.url]=[len(pwds), str(pwds)]
duration=str(datetime.timedelta(seconds=time.time() - start))
print("Number of pages counted: ", len(pornsites))
print("Duration: ", duration)
```

```python
#Bigram search with counts
stop_words = set(stopwords.words('english'))  
start = time.time()
pornsites={}
for doc in docs:
    clean = [w.strip(string.digits + string.punctuation).lower() for w in word_tokenize(doc.content) if w not in stop_words]
    bigrams=[b for b in nltk.bigrams(word_tokenize(str(' '.join(clean))))]
    match = set(bigrams).intersection(hatebig)
    pornsites[doc.url]=len(match)
duration=str(datetime.timedelta(seconds=time.time() - start))
print("Number of pages counted: ", len(pornsites))
print("Duration: ", duration)
```
